# 激活函数

* 多个全连接层如果对数据只做仿射变换（affine transform）, **多个仿射变换的叠加仍然是一个仿射变换。**
* 解决这一问题的方法是**在全连接层之间引入非线性变换。**对隐藏层的变量使用按元素运算的非线性函数进行变换，然后结果作为下一个全连接层的输入。这个非线性函数称为**激活函数（activation function）**。

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import utils as d2l
```

```python
def xy_plot(x, y, labels):
    d2l.set_figsize((6, 4))
    d2l.plt.plot(x.detach().numpy(), y.detach().numpy())
    d2l.plt.xlabel(labels[0])
    d2l.plt.ylabel(labels[1])
```

```python
# relu activation
x = torch.arange(-10.0, 8.0, 0.1, requires_grad=True)
```

## relu

$$ReLU(x) = max(x,0)$$

```python
y = x.relu()
labels = ['x', 'relu(x)']
xy_plot(x, y, labels)
```

```python
# relu grad 
v = torch.ones(y.size())
y.backward(v)
labels = ['x', 'grad_relu']
xy_plot(x, x.grad, labels)
```

## sigmoid
$$sigmoid(x)=\frac{1}{1 + exp(-x)}$$

```python
y = x.sigmoid()
labels = ['x', 'sigmoid(x)']
xy_plot(x, y, labels)
```

$$sigmoid^{'}(x)=sigmoid(x)(1-sigmoid(x))$$

```python
x.grad.zero_()
v = torch.ones(y.size())
y.backward(v)
labels = ['x', 'grad_sigmoid']
xy_plot(x, x.grad, labels)
```

## tanh
$$tanh(x) = \frac{1-exp(-2x)}{1+exp(-2x)}$$

```python
y = x.tanh()
labels = ['x', 'tanh(x)']
xy_plot(x, y, labels)
```

$$tanh^{'}(x)=1-tanh^2(x)$$

```python
x.grad.zero_()
v = torch.ones(y.size())
y.backward(v)
labels = ['x', 'grad_tanh']
xy_plot(x, x.grad, labels)
```
